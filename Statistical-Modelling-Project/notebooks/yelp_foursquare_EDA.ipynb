Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python
# Started by exploring the Foursquare API documentation. 
#There was also some useful details on how to code the API in Python on their website
#Batched the API calls to reduce the number of requests to the API

# Foursquare API

import requests
import os

def make_api_request(ll, radius, categories, fields):

    url = "https://api.foursquare.com/v3/places/search" # ll={ll}&radius={radius}&categories={categories}&sort=DISTANCE"

    headers = {"accept": "application/json",
               "Authorization": os.environ['MYKEYVARIABLE']}

    params = {
        'll': ll,
        'radius': radius,
        'categories': categories,
        'fields': fields
          }

    response = requests.get(url, params=params, headers=headers)

    return response.json()

#Make sure to append the Lat and lon of the bike stattion to the bar



```

```python
# change lat and long values back to numbers for Yelp API
bixi_stations_df['latitude'] = bixi_stations_df['latitude'].astype(float)
bixi_stations_df['longitude'] = bixi_stations_df['longitude'].astype(float)
bixi_stations_df.info()
```

```python
import requests
import os

def make_yelp_api_request(latitude, longitude, radius, term, categories, price):
    api_key = os.environ.get('YELP_API_KEY')
    
    url = "https://api.yelp.com/v3/businesses/search?"

    headers = {
         "Authorization": f"Bearer {api_key}",
        "accept": "application/json"
       }

    params = {
        'latitude': latitude,
        'longitude': longitude,
        'radius': radius, #an integer of meters, use '1000'
        'term': term, # string similar to categories, but from testing it, I see not enough places are actually. Trying 'bar'
        'categories': categories, #an array of strings of business categories. use 'bars'
        'price': price # an array of integers from 1-4
            }
    
    response = requests.get(url, params=params, headers=headers)

    return response.json()
```

```python
# Test if my API request works:
# Example parameters
latitude = 37.7749   # Example latitude (e.g., San Francisco)
longitude = -122.4194  # Example longitude
radius = 1000  # in meters
term = 'bar'
categories = 'bars'  # Example category
price = '1,2,3,4'  # Price range

# Make the API request
try:
    response = make_yelp_api_request(latitude, longitude, radius, term, categories, price)
    print(response)  # Print the response
except Exception as e:
    print(f"Error: {e}")
```
